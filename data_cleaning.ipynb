```python
import pandas as pd
import numpy as np
```

```python
#Reading our data
file_path = 'Resources/Electric_Vehicle_Population_Data.csv'
ev_df = pd.read_csv(file_path)

print(ev_df.shape)
# ev_df.head()
```

```python
# Removing all data from CSV file that is not from the state of Washington
filt = ev_df['State'] == 'WA'
clean_df = ev_df[filt]

print(clean_df.shape)
clean_df.head()
```

```python
#Saving cleaned data as CSV
clean_df.to_csv('Resources/Clean_EV_Data.csv')
```
