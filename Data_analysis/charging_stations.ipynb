```python
# Dependencies
import requests
import json
import pandas as pd

# Import the API key
from config import geoapify_key
```

```python
#Reading our data
file_path = '../Resources/Clean_EV_Data.csv'
df = pd.read_csv(file_path)

df.head()
```

```python
clean_zip = df['Postal Code']
zipcodes = list(clean_zip)
# zipcodes
```

```python
# create lists for columns in charging station dataframe
latitude_list = []
longitude_list = []
county = []
zip_code = []
name = []
```

```python
# Set the geographical coordinates for Washington
latitude = 47.7511
longitude = -120.7401

# Set the parameters for the type of place
categories = "service.vehicle.charging_station"
# conditions = "vehicle.charging_station"
radius = 339572

# Set the parameters for the type of search
filters = f"circle:{longitude},{latitude},{radius}"
bias = f"proximity:{longitude},{latitude}"
# limit = 20

# set up a parameters dictionary
params = {
    "categories":categories,
    # "conditions":conditions,
    "filter":filters,
    "bias":bias,
#     "limit":limit,
    "apiKey":geoapify_key    
}

# Set base URL
base_url = "https://api.geoapify.com/v2/places"

# run a request using our params dictionary
response = requests.get(base_url, params=params).json()

for i in range(len(response['features'])):
    latitude_list.append(response['features'][i]['properties']['lat'])
    longitude_list.append(response['features'][i]['properties']['lon'])
    county.append(response['features'][i]['properties']['county'])
#     zip_code.append(response['features'][i]['properties']['postcode'])
# name.append(response['features'][0]['properties']['name'])

print(json.dumps(response, indent=4, sort_keys=True))
```
